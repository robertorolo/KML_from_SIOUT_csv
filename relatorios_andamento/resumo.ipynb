```python
import pandas as pd
```

```python
nomes = pd.read_csv('nomes.csv')
```

```python
nomes['Nome'][0].split()
```

```python
tipo = [i.split()[0] for i in nomes['Nome']]
```

```python
import numpy as np
```

```python
tipo = np.array(tipo)
```

```python
pie_dict = {}
for s in tipo:
    ns = sum(tipo == s)
    pie_dict[s] = ns
```

```python
pie_dict
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
```

```python
plt.figure(figsize=(10,10))
plt.pie(pie_dict.values(), autopct=make_autopct(pie_dict.values()), labels=pie_dict.keys())
plt.savefig('tipo.png')
```
